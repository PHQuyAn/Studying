```r
library('tidyverse')
```

```r
library('readr')
```

```r
# url where the data is located
url <- "https://dax-cdn.cdn.appdomain.cloud/dax-airline/1.0.1/lax_to_jfk.tar.gz"
# download the file
download.file(url, destfile = "lax_to_jfk.tar.gz")
# untar the file so we can get the csv only
untar("lax_to_jfk.tar.gz", tar = "internal")
# read_csv only 
sub_airline <- read_csv("lax_to_jfk/lax_to_jfk.csv",
                     col_types = cols(
                      'DivDistance' = col_number(),
                      'DivArrDelay' = col_number()
                      ))
```

```r
filter(sub_airline, Reporting_Airline == "AA")
```

### Select: Select variables by their names

```r
select(sub_airline, c(Month, DayOfWeek, Reporting_Airline, CRSDepTime))
```

### using %>% - pipe operator

```r
sub_airline %>% 
    select(Month, DayOfWeek, Reporting_Airline, 
           CRSDepTime) %>% 
    filter(Reporting_Airline == 'AA')
```

<h3> Export data from R </h3>

```r
write_csv(sub_airline, 'sub_airline.csv')
```

<h3> R dplyr: summarize(), group_by()</h3>

```r
sub_airline %>%
    group_by(Reporting_Airline) %>%
    summarize(avg_delay = mean(ArrDelay, na.rm = TRUE))
```
